```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_df(filename):
    df = pd.read_csv(filename,skiprows = 7,sep='\t', encoding = 'utf-16',names=['lambda','abs'])
    if filename == 'hcLp1-2.csv':
        df = df[df.columns].iloc[:121]
    else:
        df = df[df.columns].iloc[:161]
    df['lambda'] = df['lambda'].astype(float)
    return df
```

```python
#ESPECTROS EN MEDIO ÁCIDO Y BÁSICO
#ácidos
df1 = read_df('hcLp1-2.csv')
df2 = read_df('hcL2.csv')
df3 = read_df('hcL3.csv')
df4 = read_df('hcL4.csv')
#bases
df5 = read_df('N1.csv')
df6 = read_df('N2.csv')
df7 = read_df('N3.csv')
df8 = read_df('N4.csv')
#pH
df9 = read_df('1.csv')
df10 = read_df('2.csv')
df11 = read_df('3.csv')
df12 = read_df('4.csv')
```

# espectros acidos defectuosos

```python


#plt.plot(df1['lambda'],df1['abs'],label='a1')
plt.plot(df2['lambda'],df2['abs'],label='a2')
plt.plot(df3['lambda'],df3['abs'],label='a3')
plt.plot(df4['lambda'],df4['abs'],label='a4')


plt.xticks(np.arange(240,390,50))
plt.xlim((240,390))
plt.xlabel("Longitud de onda [nm]")
plt.ylabel("Absorbancia")
plt.legend(frameon=False)
plt.title("Espectros de absorbancia UV-VIS en medio ácido")
plt.savefig("uv vis acidos1.jpg",dpi=200)
```
